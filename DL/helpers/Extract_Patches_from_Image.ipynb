# Extraction of patches from an image

```python
from sklearn.datasets import load_sample_images
from sklearn.feature_extraction import image
import matplotlib.pyplot as plt
```

Let us load an image and plot it:

```python
one_image = load_sample_images().images[1]
```

```python
plt.imshow(one_image)
```

```python
print('Image shape: {}'.format(one_image.shape))
```

Let us extract patches:

Method 1:

```python
patch_size = 2
stride_size = 1

patches = image.extract_patches_2d(one_image, (patch_size, patch_size))
print('Patches shape: {}'.format(patches.shape))

print(patches[1])
```

The number of patches is (427-patch_size+1) x (640-patch_size+1) = 272214

Method 2:

```python
patch_size = 2
stride_size = 1

patches=image.extract_patches(one_image,(patch_size, patch_size, 3), stride_size)
print('Patches shape: {}'.format(patches.shape))

patches=patches.reshape(-1,patch_size, patch_size, 3)
print('Patches shape: {}'.format(patches.shape))

print(patches[1])
```

As in Method 1, the number of patches is (427-patch_size+1) x (640-patch_size+1) = 272214

Method 3:

```python
patch_size = 2
stride_size = 2

patches=image.extract_patches(one_image,(patch_size, patch_size, 3), stride_size)
print('Patches shape: {}'.format(patches.shape))

patches=patches.reshape(-1,patch_size, patch_size, 3)
print('Patches shape: {}'.format(patches.shape))

print(patches[1])
```

Including a specific stride, the number of patches is (i) 427/stride_size = 213, 640/stride_size = 320, (ii) (427/stride_size) x (640/stride_size) = 68160 

# Let us resize an image:

```python
import cv2
import PIL.Image
```

```python
io=cv2.cvtColor(cv2.imread("japan.png"),cv2.COLOR_BGR2RGB)
interp_method=PIL.Image.BICUBIC
io = cv2.resize(io, dsize=(224, 224), interpolation=interp_method)
```
